```python
'''The dry-bulb temperature (DBT) is the temperature of air 
measured by a thermometer freely exposed to the air, 
but shielded from radiation and moisture. DBT is the temperature 
that is usually thought of as air temperature, and it is the true thermodynamic temperature.'''

'''The dew point is the temperature at which air is saturated with water vapor,
which is the gaseous state of water. When air has reached the dew-point temperature
at a particular pressure, the water vapor in the air is in equilibrium with liquid water,
meaning water vapor is condensing at the same rate at which liquid water is evaporating.'''
monthly_max.plot(title='dew(humidity) vs dry(thermometer) temperature',kind='hist',bins=8,alpha=0.5)
```
